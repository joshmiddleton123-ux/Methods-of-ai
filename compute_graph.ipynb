# Compute graph

While the notion of a "compute graph" will be fairly natural to people with a computer-science background, it might not be so natural to those with a Maths/Engineering background. The point of this notebook is then to introduce the simplest possible compute graph.

A compute graph is just a representation of a computation stored as data.

For instance, I could represent $2+3$ as a list, where the first element is the name of the operation, and the other elements are the arguments,

```python
["add", 2, 3]
```

Following the same idea, I could represent $3\times 4$ as,

```python
["mul", 3, 4]
```

Now, I can "nest" these expressions.  Specifically, consider $a = 2 + (3\times 4)$.  Here, $3\times 4$ is the second argument to $+$, so I can represent this expression as the following nested list:

```python
a = ["add", 2, ["mul", 3, 4]]
```

I can do this nesting arbitrarily deeply.  For instance, consider $b = (5\times 6) + (5 \times (6 \times 7))$


```python
b = ["add", ["mul", 5, 6], ["mul", 5, ["mul", 6, 7]]]
```

Critically, this code contains all the information required to manipulate, compute print etc. the expression.

```python
def compute(expr):
    if isinstance(expr, list):
        #Expr is an operation

        #First, compute the value of each argument
        arg1 = compute(expr[1])
        arg2 = compute(expr[2])

        #Then add / multiply the results
        if   expr[0] == "add":
            return arg1 + arg2
        elif expr[0] == "mul":
            return arg1 * arg2
        else:
            raise Exception(f"Unrecognised operation {expr[0]}")
    else:
        #Assume expr is just a number, and return it.
        return expr
```

Now, we can use this function to compute the value of various different compute graphs.  For instance, remember:
\begin{align}
  a = 2 + (3\times 4) = 2 + 12 = 14
\end{align}

```python
print(a)
compute(a)
```

Or,
\begin{align}
b = (5\times 6) + (5 \times (6 \times 7)) = 30 + 5 \times 42 = 30 + 210 = 240
\end{align}

```python
print(b)
compute(b)
```

The fun part is that you can do lots of other things with the compute graph.  For instance, we can covert all these nested lists into string in "standard" form (i.e. `'(2 + (3 * 4))'`).

```python
def stringify(expr):
    if isinstance(expr, list):
        #Expr is an operation

        #First, stringify each argument
        arg1 = stringify(expr[1])
        arg2 = stringify(expr[2])

        #Convert add/mul to +/*
        if   expr[0] == "add":
            func = "+"
        elif expr[0] == "mul":
            func = "*"

        #Combine everything and return the string
        return f"({arg1} {func} {arg2})"
    else:
        #Assume expr is just a number, so just use standard Python methods to convert it to a string
        return str(expr)
```

```python
print(a)
stringify(a)
```

```python
print(b)
stringify(b)
```

There are lots of difference between this and the "compute graphs" in deep learning.  Perhaps the most obvious are:
* The graph is stored in objects, rather than lists.
* The objects explicitly store the result of the operation, not just the function and arguments as here.

But this is the key idea.
